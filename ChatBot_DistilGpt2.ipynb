```python
import pandas as pd
df = pd.read_csv("hf://datasets/bitext/Bitext-customer-support-llm-chatbot-training-dataset/Bitext_Sample_Customer_Support_Training_Dataset_27K_responses-v11.csv")
```

```python
df
```

```python
df["formatted_text"] = df["instruction"] + " " + df["response"]
df
```

```python
from transformers import AutoTokenizer

MODEL_NAME = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token

# Tokenize the dataset
def tokenize_function(texts):
    return tokenizer(
        texts, truncation=True, padding="max_length", max_length=256, return_tensors="pt"
    )

# Apply tokenization
tokenized_texts = df["formatted_text"].apply(tokenize_function)
```

```python
import torch

class ChatbotDataset(torch.utils.data.Dataset):
    def __init__(self, tokenized_texts):
        self.tokenized_texts = tokenized_texts

    def __len__(self):
        return len(self.tokenized_texts)

    def __getitem__(self, idx):
        item = self.tokenized_texts.iloc[idx]
        return {
            "input_ids": item["input_ids"].squeeze(),
            "attention_mask": item["attention_mask"].squeeze(),
            "labels": item["input_ids"].squeeze(),
        }

dataset = ChatbotDataset(tokenized_texts)
```

```python
from sklearn.model_selection import train_test_split

train_texts, val_texts = train_test_split(df["formatted_text"], test_size=0.2, random_state=42)

train_encodings = train_texts.apply(tokenize_function)
val_encodings = val_texts.apply(tokenize_function)

# Create dataset objects
train_dataset = ChatbotDataset(train_encodings)
val_dataset = ChatbotDataset(val_encodings)
```

```python
from transformers import AutoModelForCausalLM, Trainer, TrainingArguments

# Load the DistilGPT-2 model
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./distilgpt2_chatbot",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    logging_dir="./logs",
)

# Initialize trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

trainer.train()
```

```python
model.save_pretrained("trained_distilgpt2_chatbot")
tokenizer.save_pretrained("trained_distilgpt2_chatbot")
```

```python
from transformers import pipeline

chatbot_pipeline = pipeline("text-generation", model="trained_distilgpt2_chatbot", tokenizer="trained_distilgpt2_chatbot")

query = "When will my order arrive?"
response = chatbot_pipeline(query, max_length=50, num_return_sequences=1)

print("Chatbot Response:", response[0]["generated_text"])
```
